```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Opening messy data
brisbane_rainfall_mess = pd.read_csv('Brisbane_end.csv', header=None, sep=';', encoding='utf-8')
# Converting string values to float & replace ',' to '.'
brisbane_rainfall = (brisbane_rainfall_mess.replace(',','.', regex=True).astype(float))
```

```python
### Pie were slices - amounts of weeks in period
```

```python
# Counting amount of weeks with needed value (1869-1970) + (1971-2021)
zero_rain_1 = (brisbane_rainfall.loc[:, 0:100] == 0).sum().sum()
zero_rain_2 = (brisbane_rainfall.loc[:, 101:] == 0).sum().sum()

min_rain_1 = ((brisbane_rainfall.loc[:, 0:100] < 10) & (brisbane_rainfall.loc[:, 0:100] != 0)).sum().sum()
min_rain_2 = ((brisbane_rainfall.loc[:, 101:] < 10) & (brisbane_rainfall.loc[:, 101:] != 0)).sum().sum()

mid_rain_1 = ((brisbane_rainfall.loc[:, 0:100] < 30) & (brisbane_rainfall.loc[:, 0:100] >= 10)).sum().sum()
mid_rain_2 = ((brisbane_rainfall.loc[:, 101:] < 30) & (brisbane_rainfall.loc[:, 101:] >= 10)).sum().sum()

big_rain_1 = ((brisbane_rainfall.loc[:, 0:100] < 100) & (brisbane_rainfall.loc[:, 0:100] >= 30)).sum().sum()
big_rain_2 = ((brisbane_rainfall.loc[:, 101:] < 100) & (brisbane_rainfall.loc[:, 101:] >= 30)).sum().sum()

flood_rain_1 = (brisbane_rainfall.loc[:, 0:100] >= 100).sum().sum()
flood_rain_2 = (brisbane_rainfall.loc[:, 101:] >= 100).sum().sum()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = '0 mm', '1-9 mm', '10-29 mm', '30-99 mm', '100+ mm'
sizes_1 = [zero_rain_1, min_rain_1, mid_rain_1, big_rain_1, flood_rain_1]
sizes_2 = [zero_rain_2, min_rain_2, mid_rain_2, big_rain_2, flood_rain_2]
explode = (0.1, 0, 0, 0.1, 0)

fig, axs = plt.subplots(1, 2, figsize=(18, 8))
colors = ['orangered', 'darkorange', 'yellow', 'skyblue', 'deepskyblue']

axs[0].pie(sizes_1, colors=colors, explode=explode, labels=labels, autopct='%1.1f%%', shadow=False, startangle=80)
axs[0].set(title='1869-1969')
axs[0].axis('equal')

axs[1].pie(sizes_2, colors=colors, explode=explode, labels=labels, autopct='%1.1f%%', shadow=False, startangle=70)
axs[1].axis('equal')
axs[1].set(title='1970-2021')

plt.show()
# Saving pie
#fig.savefig('BRISBANE_pie.png', dpi=600, bbox_inches="tight")
```

```python
flood_rain_17 = (brisbane_rainfall.loc[:, 0:16] >= 100).sum().sum()
flood_rain_34 = (brisbane_rainfall.loc[:, 17:33] >= 100).sum().sum()
flood_rain_51 = (brisbane_rainfall.loc[:, 34:50] >= 100).sum().sum()
flood_rain_68 = (brisbane_rainfall.loc[:, 51:67] >= 100).sum().sum()
flood_rain_85 = (brisbane_rainfall.loc[:, 68:84] >= 100).sum().sum()
flood_rain_102 = (brisbane_rainfall.loc[:, 85:101] >= 100).sum().sum()
flood_rain_119 = (brisbane_rainfall.loc[:, 102:118] >= 100).sum().sum()
flood_rain_136 = (brisbane_rainfall.loc[:, 119:135] >= 100).sum().sum()
flood_rain_153 = (brisbane_rainfall.loc[:, 136:] >= 100).sum().sum()
```

```python
from sklearn.linear_model import LinearRegression

years = np.array([17, 34, 51, 68, 85, 102, 119, 136, 153]).reshape((-1, 1))

zero = np.array([zero_rain_17, zero_rain_34, zero_rain_51, zero_rain_68, zero_rain_85,
                 zero_rain_102, zero_rain_119, zero_rain_136, zero_rain_153])
mini = np.array([min_rain_17, min_rain_34, min_rain_51, min_rain_68, min_rain_85,
                 min_rain_102, min_rain_119, min_rain_136, min_rain_153])
mid = np.array([mid_rain_17, mid_rain_34, mid_rain_51, mid_rain_68, mid_rain_85,
                 mid_rain_102, mid_rain_119, mid_rain_136, mid_rain_153])
big = np.array([big_rain_17, big_rain_34, big_rain_51, big_rain_68, big_rain_85,
                 big_rain_102, big_rain_119, big_rain_136, big_rain_153])
flood = np.array([flood_rain_17, flood_rain_34, flood_rain_51, flood_rain_68, flood_rain_85,
                 flood_rain_102, flood_rain_119, flood_rain_136, flood_rain_153])

model_zero = LinearRegression().fit(years, zero)
model_min = LinearRegression().fit(years, mini)
model_mid = LinearRegression().fit(years, mid)
model_big = LinearRegression().fit(years, big)
model_flood = LinearRegression().fit(years, flood)

future = np.array([17, 34, 51, 68, 85, 102, 119, 136, 153, 170, 187, 204, 221, 238, 255, 272, 289, 306]).reshape((-1, 1))
zero_future = model_zero.predict(future)
min_future = model_min.predict(future)
mid_future = model_mid.predict(future)
big_future = model_big.predict(future)
flood_future = model_flood.predict(future)
```
